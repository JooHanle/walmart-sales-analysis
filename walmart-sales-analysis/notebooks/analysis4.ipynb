### Graphiques statistiques ###

```python
import pandas as pd

# Recharger le fichier nettoyé pour garantir que nous utilisons les bonnes données
df_cleaned = pd.read_csv("C:/Users/teo11/PROJETS/walmart-sales-analysis/data/Walmart_Sales_Cleaned.csv")

# Vérifier que la colonne date est bien en datetime
df_cleaned["date"] = pd.to_datetime(df_cleaned["date"])

# Afficher les premières lignes pour confirmer que tout est correct
df_cleaned.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration générale des graphiques
plt.style.use("ggplot")

# 1. Évolution des ventes au fil du temps
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_cleaned.groupby("date")["weekly_sales"].sum().reset_index(),
             x="date", y="weekly_sales", color="blue")
plt.title("Évolution des ventes hebdomadaires")
plt.xlabel("Date")
plt.ylabel("Ventes hebdomadaires")
plt.xticks(rotation=45)
plt.show()

# 2. Moyenne des ventes par magasin
plt.figure(figsize=(12, 6))
sns.barplot(data=df_cleaned.groupby("store")["weekly_sales"].mean().reset_index(),
            x="store", y="weekly_sales", hue="store", palette="Blues_r", legend=False)
plt.title("Moyenne des ventes par magasin")
plt.xlabel("Magasin")
plt.ylabel("Ventes moyennes")
plt.show()

# 3. Impact des jours fériés sur les ventes
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_cleaned, x="holiday_flag", y="weekly_sales", hue="holiday_flag", palette="coolwarm", legend=False)
plt.title("Impact des jours fériés sur les ventes")
plt.xlabel("Jour Férié (0 = Non, 1 = Oui)")
plt.ylabel("Ventes hebdomadaires")
plt.show()

# 4. Relation entre température et ventes
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_cleaned, x="temperature", y="weekly_sales", alpha=0.5, color="green")
plt.title("Influence de la température sur les ventes")
plt.xlabel("Température (°F)")
plt.ylabel("Ventes hebdomadaires")
plt.show()

# 5. Influence du prix du carburant sur les ventes
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_cleaned, x="fuel_price", y="weekly_sales", alpha=0.5, color="red")
plt.title("Impact du prix du carburant sur les ventes")
plt.xlabel("Prix du carburant ($)")
plt.ylabel("Ventes hebdomadaires")
plt.show()
```

Voici les graphiques interactives et statiques qui mettent en évidence les premières tendances :

1. **Évolution des ventes au fil du temps :** Les ventes suivent une tendance fluctuante, avec des pics à certains moments.
2. **Comparaison des ventes par magasin :** Certains magasins enregistrent une moyenne de ventes plus élevée que d’autres.
3. **Impact des jours fériés :** Les ventes semblent varier en fonction des jours fériés.
4. **Relation température et ventes :** Il n’y a pas de tendance claire, mais certaines températures pourraient influencer les ventes.
5. **Impact du prix du carburant :** Une relation entre les prix du carburant et les ventes peut être explorée davantage.

### Analyse approfondie des chiffres et recommandations stratégiques ###
D'après l'exploration des données, voici les principales observations :

1. **Les ventes fluctuent fortement au fil du temps**, avec des pics à certaines périodes.
2. **Certains magasins performent mieux que d'autres**, indiquant un potentiel d'optimisation dans les magasins moins performants.

### 3 stratégies simples pour augmenter les ventes ###

**1. Exploiter les pics saisonniers et les jours fériés**

**Constat :** Les ventes augmentent à certaines périodes (jours fériés, fêtes).

**Action :**
Lancer des campagnes promotionnelles spéciales avant les périodes de forte demande (Noël, Thanksgiving, Black Friday).
Renforcer les stocks des produits les plus populaires durant ces périodes.
Offrir des réductions ou des offres spéciales pour inciter les achats en magasin et en ligne.

**2. Booster les ventes dans les magasins moins performants**

**Constat :** Certains magasins enregistrent des ventes plus faibles.

**Action :**
Identifier les magasins les moins performants et analyser les raisons (localisation, concurrence, stock, expérience client).
Améliorer la disposition des produits et la signalétique en magasin pour faciliter l’achat.
Mettre en place des offres promotionnelles spécifiques pour attirer plus de clients dans ces magasins.

**3. Adapter les prix et les produits aux conditions locales**

**Constat :** La température et le prix du carburant ont un impact sur les ventes.

**Action :**
Ajuster les stocks et la mise en avant des produits en fonction de la météo (exemple : boissons froides en été, vêtements chauds en hiver).
Proposer des réductions ou des programmes de fidélité lorsque le prix du carburant est élevé pour inciter les consommateurs à maintenir leurs habitudes d’achat.
Encourager les achats en ligne avec une livraison gratuite pour compenser l’impact du prix du carburant.

### En résumé ###
- Miser sur les périodes clés pour maximiser les ventes avec des promotions adaptées.
- Améliorer la performance des magasins faibles avec des offres ciblées et un meilleur aménagement.
- S’adapter aux conditions locales en modifiant les prix et l’offre produit en fonction du climat et du carburant.
